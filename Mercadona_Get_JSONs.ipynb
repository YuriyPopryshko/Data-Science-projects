# OBTAINING THE JSON FILES

### LIBRARIES

```python
import pandas as pd
import urllib.request
import json
import requests
import os
import time
```

### DOWNLOAD & UTILIZATION OF THE MASTER JSON

```python
# URL of the JSON file
url_master  = "https://tienda.mercadona.es/api/categories/"

# Perform a GET request to download the JSON file
response = requests.get(url_master )

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Save the downloaded content to a local file
    with open('categories_maestro.json', 'wb') as f:
        f.write(response.content)
    print("JSON file downloaded successfully.")

    # Load the JSON file into a dictionary
    with open('categories_maestro.json', 'r') as file:
        data = json.load(file)
    print("JSON file loaded successfully.")
else:
    print(f"Failed to download the JSON file. Status code: {response.status_code}.")
```

### DOWNLOADING JSON FOR EACH CATEGORY

```python
# Extract the IDs
ids = []
for result in data["results"]:
    for category in result["categories"]:
        ids.append(category["id"])
```

```python
base_url = "https://tienda.mercadona.es/api/categories/"
output_dir = "jsons_mercadona"
```

```python
# Download the JSON files for each category
for category_id in ids:

  
    url = f"{base_url}{category_id}"
    response = requests.get(url)
    
    
    if response.status_code == 200:
        filename = f"data_{category_id}.json"
        filepath = os.path.join(output_dir, filename)
        with open(filepath, 'wb') as f:
            f.write(response.content)
        print(f"Archivo JSON para la categoría {category_id} descargado con éxito como {filepath}.")
    else:
        print(f"No se pudo descargar el archivo JSON para la categoría {category_id}. Código de estado: {response.status_code}.")

    # Wait for a brief period between requests to not exceed the limit
    time.sleep(5)  # Wait for 5 seconds between requests
```
